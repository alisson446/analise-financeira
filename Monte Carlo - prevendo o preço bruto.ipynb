```python
import numpy as np
import matplotlib.pyplot as plt

rev_m = 170 # receita esperada em milhoes
rev_stdev = 20 # desvio padrao de 20 milhoes
iterations = 1000

rev = np.random.normal(rev_m, rev_stdev, iterations)
rev
```

```python
plt.figure(figsize = (15, 6))
plt.plot(rev)
plt.show()
```

```python
# obtendo o custo usado pra produzir o produto da empresa 
# estimando que ele usa 60% da receita pra isso, com margem de 10%
COGS = - (rev * np.random.normal(0.6, 0.1))

plt.figure(figsize = (15, 6))
plt.plot(COGS)
plt.show()
```

```python
# media do custo do produto
COGS.mean()
```

```python
# desvio padrao do custo
COGS.std()
```

```python
# lucro bruto
Gross_Profit = rev + COGS

plt.figure(figsize = (15, 6))
plt.plot(Gross_Profit)
plt.show()
```
